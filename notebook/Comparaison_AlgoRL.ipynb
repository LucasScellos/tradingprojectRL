```python
import trading_gym as tg
from stable_baselines3 import A2C, PPO, DQN
```

```python
amzn_data = tg.recuperation_donnee(start="01/01/2017", end="01/04/2022", interval="1d")
print("Nb donn√©e : ", len(amzn_data), amzn_data.head())
env_amz = tg.creation_env(df=amzn_data, frame_bound=(15, 1000))
model_A2C = tg.entrainement(env_amz, algo=A2C, timesteps=200_000)
model_PPO = tg.entrainement(env_amz, algo=PPO, timesteps=200_000)
model_DQN = tg.entrainement(env_amz, algo=DQN, timesteps=200_000)
```

```python
env_A2C, info_A2C = tg.evaluation(
    df=amzn_data, model=model_A2C, frame_bound=(1000, 1200)
)
env_PPO, info_PPO = tg.evaluation(
    df=amzn_data, model=model_PPO, frame_bound=(1000, 1200)
)
env_DQN, info_DQN = tg.evaluation(
    df=amzn_data, model=model_DQN, frame_bound=(1000, 1200)
)
for env in [env_A2C, env_PPO, env_DQN]:
    tg.graphique(env)
```

```python
def moyenne_evaluation(df, model):
    a = 0
    for _ in range(100):
        _, info = tg.evaluation(df=df, model=model, frame_bound=(1000, 1200))
        a += info["total_profit"]
    return a
```

```python
moy_A2C = moyenne_evaluation(amzn_data, model_A2C)
moy_PPO = moyenne_evaluation(amzn_data, model_PPO)
moy_DQN = moyenne_evaluation(amzn_data, model_DQN)
```
